```python
import plotly.express as px
```

```python
df = px.data.tips()
fig = px.histogram(df, x='total_bill', width=800)
fig.show()
```

```python

df = px.data.iris()
fig = px.scatter(df, x='sepal_width', y='sepal_length', color='species', symbol='species', width=800)
fig.show()
```

```python

long_df = px.data.medals_long()
fig = px.bar(long_df, x='nation', y='count', color='medal', title='Long-Form Input', width=800)
fig.show()
```

```python
df = px.data.tips()
fig = px.pie(df, values='tip', names='day', width=800)
fig.show()
```

```python
df = px.data.gapminder().query("continent == 'Oceania'")
fig = px.line(df, x='year', y='lifeExp', color='country', width=800)
fig.show()
```

```python
df = px.data.gapminder()
fig = px.scatter(df.query('year==2007'), x='gdpPercap', y='lifeExp', size='pop', color='continent', hover_name='country', log_x=True, size_max=60, width=800)
fig.show()
```
